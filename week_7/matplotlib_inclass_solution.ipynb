# Q5: subplots

```python
dx = 0.1
x = np.arange(0, 2*np.pi, dx)
f_x = np.sin(x) + np.cos(x)
df_x = np.diff(f_x)/dx
ddf_x = np.diff(df_x)/dx

plt.subplot(311)
plt.plot(x, f_x)
plt.ylabel('f(x)')

plt.subplot(312)
plt.plot(x[:-1], df_x)
plt.ylabel('df(x)')

plt.subplot(313)
plt.plot(x[:-2], ddf_x)
plt.xlabel('x')
plt.ylabel('ddf(x)')

plt.show()
```

# Q6: frequent words plotting

```python
import string
import operator

f = open("shore_leave.txt", "r")
mydict = {'a' : 0}                              # Initialize the dictionary

for line in f:
    new_line = line.replace(".", " ").replace(",", " ").split()
    for ele in new_line:
        tmp_ele = ele[0].lower() + ele[1:]
        if tmp_ele in mydict.keys():
            mydict[tmp_ele] += 1
        else:
            mydict[tmp_ele] = 1
            
sorted_dict = sorted(mydict.items(), key=operator.itemgetter(1))

# Plot top 25 most used words
words = ()
counts = []
for i in range(25):
    words +=  (sorted_dict[-(i+1)][0],)          # appending elements in to a tuple
    counts.append(sorted_dict[-(i+1)][1])        # appending elements in to a list

plt.bar(np.arange(25), counts, align='center', alpha=0.5)
# plt.xticks(np.arange(25), words)
plt.ylabel('Frequency')
plt.title('Most used words')
plt.show()
print (sorted_dict[-10: -1])
```
